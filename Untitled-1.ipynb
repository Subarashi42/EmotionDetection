# Facial Emotion Recognition for Adaptive Learning

## Project Overview

This project aims to build a facial emotion recognition system for LearnSmart Technologies, an educational technology company developing an adaptive learning platform called "EmotionLearn." The system will analyze students' facial expressions in real-time during online learning sessions to detect emotions and make immediate adjustments to the learning experience, enhancing personalization and engagement.

## Part 1: Problem Framing

### 1.1 Define the objective in business terms

- Develop a facial emotion recognition system capable of accurately classifying facial expressions into 7 emotion categories: **Angry, Disgust, Fear, Happy, Sad, Surprise, Neutral**
- Enable real-time feedback on student emotional states during online learning
- Support adaptive content delivery based on detected emotional states
- Provide analytics for teachers to understand student engagement patterns
- Improve overall learning outcomes by personalizing content based on emotional responses

---

### 1.2 How will your solution be used?

- Integrated into LearnSmart's "EmotionLearn" platform as an API
- Analyze webcam feeds at regular intervals (every 5 seconds) during online lessons
- Trigger content adaptations based on detected emotions:
  - Adjust difficulty when frustration is detected
  - Offer encouragement for negative emotions
  - Provide more challenging material when engagement is high
- Generate emotion pattern reports for teachers to identify engagement trends
- Support research on correlation between emotional states and learning outcomes

---

### 1.3 What are the current solutions/workarounds?

- Manual review of recorded sessions by teachers (time-consuming and subjective)

```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import zipfile
from PIL import Image

# For ML data preparation
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical

# Set random seeds for reproducibility
np.random.seed(42)
```

```python
def extract_fer2013_dataset(zip_filename='fer2013_data.zip', extract_path='data'):
    """Extracts the FER2013 dataset from the provided zip file."""
    # Create extraction directory if it doesn't exist
    if not os.path.exists(extract_path):
        os.makedirs(extract_path)
        print(f"Created directory: {extract_path}")
    
    # Check if the zip file exists
    if not os.path.exists(zip_filename):
        print(f"Error: {zip_filename} not found.")
        print("Please ensure the dataset zip file is in the current directory.")
        return False
    
    # Extract the contents
    try:
        print(f"Extracting {zip_filename} to {extract_path}...")
        with zipfile.ZipFile(zip_filename, 'r') as zip_ref:
            zip_ref.extractall(extract_path)
        print("Extraction complete!")
        
        # List the extracted contents
        extracted_files = os.listdir(extract_path)
        print(f"Extracted {len(extracted_files)} files/directories:")
        for item in extracted_files:
            print(f"- {item}")
        
        return True
    
    except Exception as e:
        print(f"Error extracting {zip_filename}: {e}")
        return False

# Extract the dataset
extract_success = extract_fer2013_dataset()
```

```python
def explore_fer2013_structure(data_path='data'):
    """Explores and documents the structure of the extracted FER2013 dataset."""
    print("Exploring FER2013 dataset structure...")
    
    # Define expected emotion categories
    emotion_categories = {
        0: 'Angry', 1: 'Disgust', 2: 'Fear', 
        3: 'Happy', 4: 'Sad', 5: 'Surprise', 6: 'Neutral'
    }
    
    # Check for fer2013.csv (the main dataset file)
    csv_path = os.path.join(data_path, 'fer2013.csv')
    if os.path.exists(csv_path):
        print(f"\nFound main dataset file: {csv_path}")
        
        # Load and inspect CSV file
        df = pd.read_csv(csv_path)
        print(f"Dataset shape: {df.shape}")
        print(f"Columns: {df.columns.tolist()}")
        
        # Check for expected columns
        if 'pixels' in df.columns and 'emotion' in df.columns:
            print("\nDataset format is as expected with 'pixels' and 'emotion' columns.")
            
            # Display emotion distribution
            emotion_counts = df['emotion'].value_counts().sort_index()
            print("\nEmotion distribution:")
            total_images = len(df)
            
            for emotion_idx, count in emotion_counts.items():
                emotion_name = emotion_categories.get(emotion_idx, f"Unknown ({emotion_idx})")
                percentage = (count / total_images) * 100
                print(f"- {emotion_name}: {count} images ({percentage:.2f}%)")
            
            return df
    
    # Check for alternative dataset formats (image directories)
    for subdir in ['train', 'test', 'val']:
        subdir_path = os.path.join(data_path, subdir)
        if os.path.exists(subdir_path) and os.path.isdir(subdir_path):
            print(f"\nFound {subdir} directory: {subdir_path}")
            
            # Check for emotion subdirectories
            emotion_dirs = [d for d in os.listdir(subdir_path) 
                           if os.path.isdir(os.path.join(subdir_path, d))]
            
            print(f"Contains {len(emotion_dirs)} subdirectories:")
            for emotion_dir in emotion_dirs:
                dir_path = os.path.join(subdir_path, emotion_dir)
                num_images = len([f for f in os.listdir(dir_path) 
                                 if f.endswith(('.jpg', '.png', '.jpeg'))])
                print(f"- {emotion_dir}: {num_images} images")
    
    print("\nDataset exploration complete.")
    return None

# Explore the dataset structure
fer_df = explore_fer2013_structure()
```

```python
def load_fer2013_data(data_path='data', df=None):
    """Loads the FER2013 dataset into memory."""
    # Define emotion mapping
    emotion_map = {
        0: 'Angry', 1: 'Disgust', 2: 'Fear', 
        3: 'Happy', 4: 'Sad', 5: 'Surprise', 6: 'Neutral'
    }
    
    # First, try loading from CSV
    csv_path = os.path.join(data_path, 'fer2013.csv')
    if df is None and os.path.exists(csv_path):
        print(f"Loading data from {csv_path}...")
        df = pd.read_csv(csv_path)
    
    if df is not None and 'pixels' in df.columns and 'emotion' in df.columns:
        print("Processing data from DataFrame...")
        
        # Parse string pixel values into numpy arrays
        print("Converting pixel strings to image arrays...")
        X = []
        
        # Simple progress tracking
        total_rows = len(df)
        print_intervals = [int(total_rows * i / 10) for i in range(1, 11)]
        
        for i, pixel_str in enumerate(df['pixels']):
            # Simple progress reporting
            if i in print_intervals:
                print(f"Progress: {i}/{total_rows} images ({i/total_rows*100:.1f}%)")
                
            pixels = [int(p) for p in pixel_str.split()]
            X.append(np.array(pixels).reshape(48, 48))
        
        X = np.array(X)
        y = df['emotion'].values
        
        print(f"Loaded {len(X)} images with shape {X[0].shape}")
        return X, y, emotion_map
    
    # If CSV loading fails, try loading from image directories
    print("CSV loading failed, checking for image directories...")
    
    X = []
    y = []
    total_images = 0
    
    # First count total images
    for split in ['train', 'test', 'val']:
        split_dir = os.path.join(data_path, split)
        if not os.path.exists(split_dir):
            continue
        
        for emotion_idx, emotion_name in emotion_map.items():
            # Try different directory naming conventions
            for dir_name in [emotion_name.lower(), str(emotion_idx)]:
                emotion_dir = os.path.join(split_dir, dir_name)
                if os.path.exists(emotion_dir):
                    total_images += len([f for f in os.listdir(emotion_dir) 
                                        if f.lower().endswith(('.jpg', '.jpeg', '.png'))])
    
    print(f"Found {total_images} images in directories. Loading...")
    processed_images = 0
    print_intervals = [int(total_images * i / 10) for i in range(1, 11)]
    
    # Load images from directories
    for split in ['train', 'test', 'val']:
        split_dir = os.path.join(data_path, split)
        if not os.path.exists(split_dir):
            continue
        
        print(f"Loading images from {split_dir}...")
        
        for emotion_idx, emotion_name in emotion_map.items():
            # Try different directory naming conventions
            for dir_name in [emotion_name.lower(), str(emotion_idx)]:
                emotion_dir = os.path.join(split_dir, dir_name)
                if not os.path.exists(emotion_dir):
                    continue
                
                print(f"Loading {emotion_name} images...")
                for img_file in os.listdir(emotion_dir):
                    if img_file.lower().endswith(('.jpg', '.jpeg', '.png')):
                        img_path = os.path.join(emotion_dir, img_file)
                        img = Image.open(img_path).convert('L')  # Convert to grayscale
                        img_resized = img.resize((48, 48))
                        img_array = np.array(img_resized)
                        X.append(img_array)
                        y.append(emotion_idx)
                        
                        processed_images += 1
                        if processed_images in print_intervals:
                            print(f"Progress: {processed_images}/{total_images} images ({processed_images/total_images*100:.1f}%)")
    
    if len(X) > 0:
        X = np.array(X)
        y = np.array(y)
        print(f"Loaded {len(X)} images with shape {X[0].shape}")
        return X, y, emotion_map
    
    print("Failed to load dataset. Please check the dataset structure.")
    return None, None, emotion_map

# Load the dataset
X, y, emotion_map = load_fer2013_data(df=fer_df)
```

```python
def visualize_emotion_samples(X, y, emotion_map, samples_per_emotion=3):
    """Visualizes sample images from each emotion category."""
    if X is None or y is None:
        print("No data available to visualize.")
        return
    
    # Get all unique emotion labels
    unique_emotions = sorted(np.unique(y))
    num_emotions = len(unique_emotions)
    
    # Create a figure with subplots
    fig, axes = plt.subplots(num_emotions, samples_per_emotion, 
                            figsize=(samples_per_emotion*3, num_emotions*2.5))
    
    # For each emotion, select random samples
    for i, emotion_idx in enumerate(unique_emotions):
        # Get indices of all images with this emotion
        emotion_indices = np.where(y == emotion_idx)[0]
        
        # Select random samples
        if len(emotion_indices) < samples_per_emotion:
            sample_indices = emotion_indices
        else:
            sample_indices = np.random.choice(emotion_indices, samples_per_emotion, replace=False)
        
        # Display each sample
        for j, idx in enumerate(sample_indices):
            ax = axes[i, j] if num_emotions > 1 else axes[j]
            ax.imshow(X[idx], cmap='gray')
            
            if j == 0:  # Only add emotion label to first sample in row
                ax.set_ylabel(emotion_map[emotion_idx], fontsize=12)
            
            ax.set_xticks([])
            ax.set_yticks([])
    
    plt.tight_layout()
    plt.suptitle('Sample Images from Each Emotion Category', y=1.02, fontsize=16)
    plt.show()

# Visualize sample images
visualize_emotion_samples(X, y, emotion_map)
```

```python
def split_dataset(X, y, test_size=0.1, val_size=0.1):
    """Splits the dataset into training, validation, and test sets."""
    if X is None or y is None:
        print("No data available to split.")
        return None, None, None, None, None, None
    
    # First split: separate test set
    X_temp, X_test, y_temp, y_test = train_test_split(
        X, y, test_size=test_size, random_state=42, stratify=y
    )
    
    # Second split: separate validation set from remaining data
    val_size_adjusted = val_size / (1 - test_size)
    X_train, X_val, y_train, y_val = train_test_split(
        X_temp, y_temp, test_size=val_size_adjusted, random_state=42, stratify=y_temp
    )
    
    print("Dataset split complete:")
    print(f"Training set: {len(X_train)} images ({len(X_train)/len(X)*100:.1f}%)")
    print(f"Validation set: {len(X_val)} images ({len(X_val)/len(X)*100:.1f}%)")
    print(f"Test set: {len(X_test)} images ({len(X_test)/len(X)*100:.1f}%)")
    
    # Verify class distribution is preserved
    def print_distribution(name, labels):
        print(f"\n{name} set emotion distribution:")
        unique, counts = np.unique(labels, return_counts=True)
        for u, c in zip(unique, counts):
            print(f"- {emotion_map[u]}: {c} images ({c/len(labels)*100:.1f}%)")
    
    print_distribution("Training", y_train)
    print_distribution("Validation", y_val)
    print_distribution("Test", y_test)
    
    return X_train, X_val, X_test, y_train, y_val, y_test

# Split the dataset
X_train, X_val, X_test, y_train, y_val, y_test = split_dataset(X, y)
```

```python
def save_test_set(X_test, y_test, save_dir='data/test_set'):
    """Saves the test set separately to avoid data leakage."""
    if X_test is None or y_test is None:
        return
    
    # Create directory if it doesn't exist
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    
    # Save test data
    np.save(os.path.join(save_dir, 'X_test.npy'), X_test)
    np.save(os.path.join(save_dir, 'y_test.npy'), y_test)
    
    print(f"\nTest set saved to {save_dir}")
    print("This test set should only be used for final model evaluation.")

# Save the test set
save_test_set(X_test, y_test)
```

```python
def preprocess_data(X_train, X_val, X_test, y_train, y_val, y_test):
    """Performs basic preprocessing on the data for modeling."""
    if X_train is None or y_train is None:
        print("No data available to preprocess.")
        return None, None, None, None, None, None
    
    # Normalize pixel values to [0, 1]
    X_train_norm = X_train.astype('float32') / 255.0
    X_val_norm = X_val.astype('float32') / 255.0
    X_test_norm = X_test.astype('float32') / 255.0
    
    # Add channel dimension for CNN models (height, width, channels)
    X_train_cnn = X_train_norm.reshape(X_train_norm.shape[0], 48, 48, 1)
    X_val_cnn = X_val_norm.reshape(X_val_norm.shape[0], 48, 48, 1)
    X_test_cnn = X_test_norm.reshape(X_test_norm.shape[0], 48, 48, 1)
    
    # One-hot encode the labels for categorical crossentropy loss
    num_classes = len(np.unique(y_train))
    y_train_onehot = to_categorical(y_train, num_classes)
    y_val_onehot = to_categorical(y_val, num_classes)
    y_test_onehot = to_categorical(y_test, num_classes)
    
    print("Preprocessing complete:")
    print(f"X_train shape: {X_train_cnn.shape}")
    print(f"X_val shape: {X_val_cnn.shape}")
    print(f"X_test shape: {X_test_cnn.shape}")
    print(f"y_train shape: {y_train_onehot.shape}")
    
    return X_train_cnn, X_val_cnn, X_test_cnn, y_train_onehot, y_val_onehot, y_test_onehot

# Preprocess the data
X_train_cnn, X_val_cnn, X_test_cnn, y_train_onehot, y_val_onehot, y_test_onehot = preprocess_data(
    X_train, X_val, X_test, y_train, y_val, y_test
)
```

```python
def save_processed_data(X_train, X_val, y_train, y_val, save_dir='data/processed'):
    """Saves the preprocessed data for future use."""
    if X_train is None or y_train is None:
        return
    
    # Create directory if it doesn't exist
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    
    # Save preprocessed data
    np.save(os.path.join(save_dir, 'X_train_cnn.npy'), X_train)
    np.save(os.path.join(save_dir, 'X_val_cnn.npy'), X_val)
    np.save(os.path.join(save_dir, 'y_train_onehot.npy'), y_train)
    np.save(os.path.join(save_dir, 'y_val_onehot.npy'), y_val)
    
    print(f"\nPreprocessed data saved to {save_dir}")

# Save the preprocessed data
save_processed_data(X_train_cnn, X_val_cnn, y_train_onehot, y_val_onehot)
```

### 1.2 Calculate Basic Statistics for Each Class

```python
def calculate_class_statistics(X_train, y_train, emotion_map):
    """Calculate basic statistics for each emotion category."""
    print("Class Statistics:\n")
    
    # Get all unique emotion labels
    unique_emotions = sorted(np.unique(y_train))
    
    # Create a table of statistics
    stats = []
    
    for emotion_idx in unique_emotions:
        # Get all images for this emotion
        emotion_images = X_train[y_train == emotion_idx]
        
        # Calculate statistics
        mean_intensity = np.mean(emotion_images)
        std_intensity = np.std(emotion_images)
        min_intensity = np.min(emotion_images)
        max_intensity = np.max(emotion_images)
        median_intensity = np.median(emotion_images)
        
        # Store statistics
        stats.append({
            'Emotion': emotion_map[emotion_idx],
            'Count': len(emotion_images),
            'Mean Intensity': mean_intensity,
            'Std Intensity': std_intensity,
            'Min Intensity': min_intensity,
            'Median Intensity': median_intensity,
            'Max Intensity': max_intensity
        })
    
    # Create DataFrame for better visualization
    stats_df = pd.DataFrame(stats)
    print(stats_df)
    
    return stats_df

# Calculate statistics for each class
class_stats = calculate_class_statistics(X_train, y_train, emotion_map)
```

### 1.7. Explore Correlation Between Emotions

```python
def analyze_emotion_correlation(X_train, y_train, emotion_map):
    """Analyzes correlation between average faces of different emotions."""
    # Get all unique emotion labels
    unique_emotions = sorted(np.unique(y_train))
    num_emotions = len(unique_emotions)
    
    # Calculate average face for each emotion
    avg_faces = {}
    for emotion_idx in unique_emotions:
        emotion_images = X_train[y_train == emotion_idx]
        avg_faces[emotion_idx] = np.mean(emotion_images, axis=0).flatten()
    
    # Calculate correlation matrix
    corr_matrix = np.zeros((num_emotions, num_emotions))
    
    for i, emotion1 in enumerate(unique_emotions):
        for j, emotion2 in enumerate(unique_emotions):
            corr = np.corrcoef(avg_faces[emotion1], avg_faces[emotion2])[0, 1]
            corr_matrix[i, j] = corr
    
    # Create correlation DataFrame
    emotion_names = [emotion_map[idx] for idx in unique_emotions]
    corr_df = pd.DataFrame(corr_matrix, index=emotion_names, columns=emotion_names)
    
    # Visualize correlation matrix
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_df, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.title('Correlation Between Average Faces of Different Emotions')
    plt.tight_layout()
    plt.show()
    
    return corr_df

# Analyze correlation between emotions
emotion_corr = analyze_emotion_correlation(X_train, y_train, emotion_map)
```

### 1.8. Identify Hard-to-Distinguish Emotion Pairs

```python
def identify_challenging_pairs(emotion_corr, emotion_map):
    """Identifies potentially challenging emotion pairs for classification."""
    # Create a copy of correlation matrix to avoid modifying the original
    corr_matrix = emotion_corr.copy()
    
    # Set diagonal to -1 (to ignore self-correlations)
    np.fill_diagonal(corr_matrix.values, -1)
    
    # Find top 5 most correlated pairs
    n_pairs = 5
    challenging_pairs = []
    
    for _ in range(n_pairs):
        # Find the maximum correlation
        max_idx = corr_matrix.values.argmax()
        i, j = np.unravel_index(max_idx, corr_matrix.shape)
        
        # Get emotion names
        emotion1 = corr_matrix.index[i]
        emotion2 = corr_matrix.columns[j]
        
        # Get the correlation value
        corr_value = corr_matrix.iloc[i, j]
        
        # Add to list
        challenging_pairs.append((emotion1, emotion2, corr_value))
        
        # Set this pair's correlation to -1 to ignore in next iteration
        corr_matrix.iloc[i, j] = -1
        corr_matrix.iloc[j, i] = -1
    
    # Print results
    print("Top 5 potentially challenging emotion pairs to distinguish:")
    for emotion1, emotion2, corr in challenging_pairs:
        print(f"- {emotion1} vs {emotion2}: Correlation = {corr:.4f}")
    
    return challenging_pairs

# Identify challenging emotion pairs
challenging_pairs = identify_challenging_pairs(emotion_corr, emotion_map)
```

### 2.2 Examine Pixel Importance Using Heatmaps

```python
def compute_pixel_importance(X_train, y_train):
    """
    Computes pixel importance for emotion discrimination using variance analysis.
    """
    # Get unique emotion labels
    unique_emotions = sorted(np.unique(y_train))
    num_emotions = len(unique_emotions)
    
    # Initialize arrays for mean faces and overall mean
    mean_faces = np.zeros((num_emotions, X_train.shape[1], X_train.shape[2]))
    overall_mean = np.mean(X_train, axis=0)
    
    # Compute mean face for each emotion
    for i, emotion_idx in enumerate(unique_emotions):
        emotion_images = X_train[y_train == emotion_idx]
        mean_faces[i] = np.mean(emotion_images, axis=0)
    
    # Compute between-class variance for each pixel
    between_var = np.zeros_like(overall_mean)
    for i in range(num_emotions):
        between_var += (mean_faces[i] - overall_mean) ** 2
    between_var /= num_emotions
    
    # Compute within-class variance for each pixel
    within_var = np.zeros_like(overall_mean)
    for i, emotion_idx in enumerate(unique_emotions):
        emotion_images = X_train[y_train == emotion_idx]
        for img in emotion_images:
            within_var += (img - mean_faces[i]) ** 2
    within_var /= len(X_train)
    
    # Compute Fisher's discriminant score (between-class / within-class variance)
    # Add small constant to avoid division by zero
    fisher_score = between_var / (within_var + 1e-10)
    
    return fisher_score, between_var, within_var

def visualize_pixel_importance(X_train, y_train, emotion_map):
    """
    Visualizes pixel importance for emotion discrimination.
    """
    # Compute pixel importance
    fisher_score, between_var, within_var = compute_pixel_importance(X_train, y_train)
    
    # Visualize results
    plt.figure(figsize=(15, 5))
    
    # Plot between-class variance
    plt.subplot(1, 3, 1)
    plt.imshow(between_var, cmap='hot')
    plt.title('Between-class Variance')
    plt.colorbar()
    plt.axis('off')
    
    # Plot within-class variance
    plt.subplot(1, 3, 2)
    plt.imshow(within_var, cmap='hot')
    plt.title('Within-class Variance')
    plt.colorbar()
    plt.axis('off')
    
    # Plot Fisher's discriminant score
    plt.subplot(1, 3, 3)
    plt.imshow(fisher_score, cmap='hot')
    plt.title('Pixel Importance (Fisher Score)')
    plt.colorbar()
    plt.axis('off')
    
    plt.suptitle('Pixel Importance Analysis for Emotion Recognition', y=0.98, fontsize=16)
    plt.tight_layout()
    plt.subplots_adjust(top=0.85)
    plt.show()
    
    # Overlay importance map on average face
    plt.figure(figsize=(10, 8))
    
    # Overall average face
    overall_mean = np.mean(X_train, axis=0)
    
    # Normalize Fisher score for visualization
    normalized_fisher = (fisher_score - fisher_score.min()) / (fisher_score.max() - fisher_score.min())
    
    # Create a color overlay
    plt.imshow(overall_mean, cmap='gray', alpha=1.0)
    plt.imshow(normalized_fisher, cmap='hot', alpha=0.6)
    plt.title('Important Facial Regions for Emotion Recognition')
    plt.colorbar(label='Relative Importance')
    plt.axis('off')
    plt.tight_layout()
    plt.show()
    
    return fisher_score

# Visualize pixel importance
pixel_importance = visualize_pixel_importance(X_train, y_train, emotion_map)
```

### Extract Facial Regions of Interest

```python
def extract_facial_regions(X_train, y_train, emotion_map):
    """
    Extracts key facial regions (eyes, mouth, etc.) based on approximate positions.
    """
    # Image dimensions
    height, width = X_train[0].shape
    
    # Define regions of interest (approximate)
    regions = {
        'left_eye': (int(height*0.2), int(height*0.45), int(width*0.2), int(width*0.45)),
        'right_eye': (int(height*0.2), int(height*0.45), int(width*0.55), int(width*0.8)),
        'mouth': (int(height*0.6), int(height*0.85), int(width*0.3), int(width*0.7)),
        'forehead': (int(height*0.05), int(height*0.2), int(width*0.3), int(width*0.7)),
        'nose': (int(height*0.45), int(height*0.6), int(width*0.4), int(width*0.6))
    }
    
    # Extract regions and visualize examples
    plt.figure(figsize=(15, 10))
    
    # Select a sample image from each emotion
    for i, emotion_idx in enumerate(sorted(np.unique(y_train))):
        # Get a sample image for this emotion
        emotion_indices = np.where(y_train == emotion_idx)[0]
        sample_idx = np.random.choice(emotion_indices)
        sample_image = X_train[sample_idx]
        
        # Plot the full face
        plt.subplot(len(emotion_map), len(regions)+1, i*(len(regions)+1)+1)
        plt.imshow(sample_image, cmap='gray')
        plt.title(f'{emotion_map[emotion_idx]}')
        plt.axis('off')
        
        # Plot each region
        for j, (region_name, (y_start, y_end, x_start, x_end)) in enumerate(regions.items()):
            region_img = sample_image[y_start:y_end, x_start:x_end]
            plt.subplot(len(emotion_map), len(regions)+1, i*(len(regions)+1)+j+2)
            plt.imshow(region_img, cmap='gray')
            if i == 0:  # Only add region title to first row
                plt.title(region_name)
            plt.axis('off')
    
    plt.suptitle('Facial Regions by Emotion', y=0.98, fontsize=16)
    plt.tight_layout()
    plt.subplots_adjust(top=0.95)
    plt.show()
    
    # Extract features from each region
    region_features = {region: [] for region in regions}
    
    for i in range(len(X_train)):
        for region_name, (y_start, y_end, x_start, x_end) in regions.items():
            region_img = X_train[i][y_start:y_end, x_start:x_end]
            
            # Compute features for this region
            features = {
                'mean': np.mean(region_img),
                'std': np.std(region_img),
                'min': np.min(region_img),
                'max': np.max(region_img),
                'range': np.max(region_img) - np.min(region_img)
            }
            
            region_features[region_name].append(features)
    
    return regions, region_features

# Extract facial regions
regions, region_features = extract_facial_regions(X_train, y_train, emotion_map)
```

### Compute Texture-Based Features

```python
from skimage.feature import local_binary_pattern, hog
from skimage import exposure

def extract_texture_features(X_train, y_train, emotion_map, sample_size=10):
    """
    Extracts texture-based features using HOG and LBP.
    """
    # LBP parameters
    radius = 3
    n_points = 8 * radius
    
    # HOG parameters
    orientations = 8
    pixels_per_cell = (8, 8)
    cells_per_block = (2, 2)
    
    # Select sample images
    sample_indices = []
    for emotion_idx in sorted(np.unique(y_train)):
        emotion_indices = np.where(y_train == emotion_idx)[0]
        selected = np.random.choice(emotion_indices, min(sample_size, len(emotion_indices)), replace=False)
        sample_indices.extend(selected)
    
    # Sort by emotion for better visualization
    sample_indices = sorted(sample_indices, key=lambda x: y_train[x])
    
    # Create figure
    plt.figure(figsize=(15, len(sample_indices)*2))
    
    # For each sample image
    for i, idx in enumerate(sample_indices):
        img = X_train[idx]
        emotion = emotion_map[y_train[idx]]
        
        # Compute LBP
        lbp = local_binary_pattern(img, n_points, radius, method='uniform')
        
        # Compute HOG
        hog_features, hog_image = hog(
            img, 
            orientations=orientations, 
            pixels_per_cell=pixels_per_cell,
            cells_per_block=cells_per_block, 
            visualize=True, 
            block_norm='L2-Hys'
        )
        
        # Enhance HOG visualization
        hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))
        
        # Plot original image
        plt.subplot(len(sample_indices), 3, i*3 + 1)
        plt.imshow(img, cmap='gray')
        plt.title(f'{emotion} - Original')
        plt.axis('off')
        
        # Plot LBP
        plt.subplot(len(sample_indices), 3, i*3 + 2)
        plt.imshow(lbp, cmap='gray')
        plt.title(f'{emotion} - LBP')
        plt.axis('off')
        
        # Plot HOG
        plt.subplot(len(sample_indices), 3, i*3 + 3)
        plt.imshow(hog_image_rescaled, cmap='gray')
        plt.title(f'{emotion} - HOG')
        plt.axis('off')
    
    plt.suptitle('Texture Features for Emotion Recognition', y=0.99, fontsize=16)
    plt.tight_layout()
    plt.subplots_adjust(top=0.95)
    plt.show()
    
    return None  # We're just visualizing here

# Extract and visualize texture features
extract_texture_features(X_train, y_train, emotion_map)
```

### Analyze Emotion Transitions and Gradients

```python
from sklearn.decomposition import PCA
from scipy.spatial import Delaunay
import matplotlib.path as mpath

def analyze_emotion_transitions(X_train, y_train, emotion_map):
    """
    Analyzes transitions between emotions in feature space.
    """
    # Flatten images
    X_flat = X_train.reshape(X_train.shape[0], -1)
    
    # Apply PCA for dimensionality reduction
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X_flat)
    
    # Create figure
    plt.figure(figsize=(12, 10))
    
    # Get unique emotions
    unique_emotions = sorted(np.unique(y_train))
    
    # Plot each emotion
    emotion_points = {}
    for emotion_idx in unique_emotions:
        # Get points for this emotion
        mask = y_train == emotion_idx
        points = X_pca[mask]
        
        # Store points
        emotion_points[emotion_idx] = points
        
        # Plot points
        plt.scatter(
            points[:, 0], 
            points[:, 1],
            s=30,
            alpha=0.7,
            label=emotion_map[emotion_idx]
        )
    
    # Calculate and plot emotion centroids
    centroids = {}
    for emotion_idx, points in emotion_points.items():
        centroid = np.mean(points, axis=0)
        centroids[emotion_idx] = centroid
        
        plt.scatter(
            centroid[0],
            centroid[1],
            s=200,
            marker='*',
            color='black',
            edgecolor='w',
            linewidth=1.5
        )
        
        plt.annotate(
            emotion_map[emotion_idx],
            xy=(centroid[0], centroid[1]),
            xytext=(centroid[0] + 0.05, centroid[1] + 0.05),
            fontsize=12,
            fontweight='bold'
        )
    
    # Connect centroids to show emotion transitions
    edges = []
    for i, emotion1 in enumerate(unique_emotions):
        for emotion2 in unique_emotions[i+1:]:
            # Get centroids
            centroid1 = centroids[emotion1]
            centroid2 = centroids[emotion2]
            
            # Calculate distance
            dist = np.sqrt(np.sum((centroid1 - centroid2)**2))
            
            # Store edge
            edges.append((emotion1, emotion2, dist))
    
    # Sort edges by distance
    edges.sort(key=lambda x: x[2])
    
    # Get top connections (minimum spanning tree + some extras)
    n_connections = len(unique_emotions) + 2  # MST + a few extras
    n_connections = min(n_connections, len(edges))
    
    # Plot connections
    for i in range(n_connections):
        emotion1, emotion2, _ = edges[i]
        centroid1 = centroids[emotion1]
        centroid2 = centroids[emotion2]
        
        plt.plot(
            [centroid1[0], centroid2[0]],
            [centroid1[1], centroid2[1]],
            'k--',
            alpha=0.5
        )
        
        # Add midpoint label with distance
        mid_x = (centroid1[0] + centroid2[0]) / 2
        mid_y = (centroid1[1] + centroid2[1]) / 2
        
        plt.annotate(
            f"{emotion_map[emotion1]}-{emotion_map[emotion2]}",
            xy=(mid_x, mid_y),
            xytext=(mid_x, mid_y),
            fontsize=8,
            ha='center',
            va='center',
            bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.7)
        )
    
    plt.title('Emotion Transitions in PCA Space', fontsize=16)
    plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}%)', fontsize=12)
    plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}%)', fontsize=12)
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    return centroids, edges

# Analyze emotion transitions
emotion_centroids, emotion_edges = analyze_emotion_transitions(X_train, y_train, emotion_map)
```

### Generate Statistical Features from Pixel Distributions

```python
def generate_statistical_features(X_train, y_train, emotion_map):
    """
    Generates statistical features from pixel distributions for each emotion.
    """
    # Get unique emotions
    unique_emotions = sorted(np.unique(y_train))
    
    # Initialize feature dictionary
    features = {emotion_idx: {} for emotion_idx in unique_emotions}
    
    # Calculate statistical features for each emotion
    for emotion_idx in unique_emotions:
        # Get images for this emotion
        emotion_images = X_train[y_train == emotion_idx]
        
        # Calculate features
        features[emotion_idx]['mean'] = np.mean(emotion_images, axis=(1, 2))
        features[emotion_idx]['std'] = np.std(emotion_images, axis=(1, 2))
        features[emotion_idx]['median'] = np.median(emotion_images, axis=(1, 2))
        features[emotion_idx]['min'] = np.min(emotion_images, axis=(1, 2))
        features[emotion_idx]['max'] = np.max(emotion_images, axis=(1, 2))
        features[emotion_idx]['range'] = features[emotion_idx]['max'] - features[emotion_idx]['min']
        features[emotion_idx]['q25'] = np.percentile(emotion_images, 25, axis=(1, 2))
        features[emotion_idx]['q75'] = np.percentile(emotion_images, 75, axis=(1, 2))
        features[emotion_idx]['iqr'] = features[emotion_idx]['q75'] - features[emotion_idx]['q25']
    
    # Visualize feature distributions
    feature_names = list(features[unique_emotions[0]].keys())
    
    # Create figure
    plt.figure(figsize=(20, 15))
    
    # For each feature
    for i, feature_name in enumerate(feature_names):
        plt.subplot(3, 3, i+1)
        
        # For each emotion
        for emotion_idx in unique_emotions:
            # Get feature values
            values = features[emotion_idx][feature_name]
            
            # Plot distribution
            sns.kdeplot(values, label=emotion_map[emotion_idx])
        
        plt.title(f'{feature_name.capitalize()} Distribution by Emotion')
        plt.xlabel(feature_name.capitalize())
        plt.ylabel('Density')
        plt.legend()
        plt.grid(alpha=0.3)
    
    plt.suptitle('Statistical Feature Distributions by Emotion', y=0.98, fontsize=16)
    plt.tight_layout()
    plt.subplots_adjust(top=0.93)
    plt.show()
    
    # Analyze feature importance
    # Calculate mean and std for each feature across emotions
    feature_stats = {}
    for feature_name in feature_names:
        feature_stats[feature_name] = {
            'means': [],
            'stds': []
        }
        
        for emotion_idx in unique_emotions:
            feature_stats[feature_name]['means'].append(np.mean(features[emotion_idx][feature_name]))
            feature_stats[feature_name]['stds'].append(np.std(features[emotion_idx][feature_name]))
    
    # Calculate feature importance based on variance between emotion means
    feature_importance = {}
    for feature_name in feature_names:
        # Calculate variance between emotion means
        variance = np.var(feature_stats[feature_name]['means'])
        
        # Calculate average within-class variance
        avg_variance = np.mean([std**2 for std in feature_stats[feature_name]['stds']])
        
        # Calculate Fisher's score: between-class / within-class variance
        fisher_score = variance / (avg_variance + 1e-10)
        
        # Store importance
        feature_importance[feature_name] = fisher_score
    
    # Visualize feature importance
    plt.figure(figsize=(12, 6))
    
    # Sort features by importance
    sorted_features = sorted(feature_importance.items(), key=lambda x: x[1], reverse=True)
    feature_names = [item[0] for item in sorted_features]
    importance_values = [item[1] for item in sorted_features]
    
    # Plot
    sns.barplot(x=feature_names, y=importance_values)
    plt.title('Statistical Feature Importance for Emotion Discrimination')
    plt.xlabel('Feature')
    plt.ylabel('Importance (Fisher Score)')
    plt.xticks(rotation=45)
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    return features, feature_importance

# Generate and analyze statistical features
stat_features, feature_importance = generate_statistical_features(X_train, y_train, emotion_map)
```

###  Edge Detection and Gradient Analysis

```python
from scipy import ndimage
from skimage import feature, filters

def analyze_edge_patterns(X_train, y_train, emotion_map, sample_size=3):
    """
    Analyzes edge patterns in facial images across emotions.
    """
    # Get unique emotions
    unique_emotions = sorted(np.unique(y_train))
    
    # Create figure
    plt.figure(figsize=(15, len(unique_emotions)*2.5))
    
    # For each emotion
    for i, emotion_idx in enumerate(unique_emotions):
        # Get images for this emotion
        emotion_indices = np.where(y_train == emotion_idx)[0]
        
        # Select sample images
        selected_indices = np.random.choice(emotion_indices, min(sample_size, len(emotion_indices)), replace=False)
        
        # For each sample
        for j, idx in enumerate(selected_indices):
            img = X_train[idx]
            
            # Calculate Sobel gradients
            sobel_h = ndimage.sobel(img, axis=0)
            sobel_v = ndimage.sobel(img, axis=1)
            magnitude = np.sqrt(sobel_h**2 + sobel_v**2)
            
            # Calculate Canny edges
            canny = feature.canny(img, sigma=1.0)
            
            # Plot original image
            plt.subplot(len(unique_emotions), sample_size*3, i*sample_size*3 + j*3 + 1)
            plt.imshow(img, cmap='gray')
            if j == 0:
                plt.ylabel(emotion_map[emotion_idx], fontsize=12)
            if i == 0:
                plt.title('Original')
            plt.axis('off')
            
            # Plot gradient magnitude
            plt.subplot(len(unique_emotions), sample_size*3, i*sample_size*3 + j*3 + 2)
            plt.imshow(magnitude, cmap='inferno')
            if i == 0:
                plt.title('Gradient Magnitude')
            plt.axis('off')
            
            # Plot Canny edges
            plt.subplot(len(unique_emotions), sample_size*3, i*sample_size*3 + j*3 + 3)
            plt.imshow(canny, cmap='gray')
            if i == 0:
                plt.title('Canny Edges')
            plt.axis('off')
    
    plt.suptitle('Edge Patterns by Emotion', y=0.98, fontsize=16)
    plt.tight_layout()
    plt.subplots_adjust(top=0.95)
    plt.show()
    
    # Compute average edge patterns for each emotion
    plt.figure(figsize=(15, 10))
    
    for i, emotion_idx in enumerate(unique_emotions):
        # Get images for this emotion
        emotion_images = X_train[y_train == emotion_idx]
        
        # Compute average image
        avg_img = np.mean(emotion_images, axis=0)
        
        # Compute average gradient magnitude
        avg_magnitude = np.zeros_like(avg_img)
        avg_canny = np.zeros_like(avg_img)
        
        for img in emotion_images:
            # Calculate Sobel gradients
            sobel_h = ndimage.sobel(img, axis=0)
            sobel_v = ndimage.sobel(img, axis=1)
            magnitude = np.sqrt(sobel_h**2 + sobel_v**2)
            
            # Calculate Canny edges
            canny = feature.canny(img, sigma=1.0)
            
            # Accumulate
            avg_magnitude += magnitude
            avg_canny += canny
        
        # Normalize
        avg_magnitude /= len(emotion_images)
        avg_canny /= len(emotion_images)
        
        # Plot average image
        plt.subplot(len(unique_emotions), 3, i*3 + 1)
        plt.imshow(avg_img, cmap='gray')
        plt.title(f'Avg {emotion_map[emotion_idx]}')
        plt.axis('off')
        
        # Plot average gradient magnitude
        plt.subplot(len(unique_emotions), 3, i*3 + 2)
        plt.imshow(avg_magnitude, cmap='inferno')
        plt.title(f'Avg Gradient')
        plt.axis('off')
        
        # Plot average Canny edges
        plt.subplot(len(unique_emotions), 3, i*3 + 3)
        plt.imshow(avg_canny, cmap='gray')
        plt.title(f'Avg Edges')
        plt.axis('off')
    
    plt.suptitle('Average Edge Patterns by Emotion', y=0.98, fontsize=16)
    plt.tight_layout()
    plt.subplots_adjust(top=0.95)
    plt.show()
    
    return None

# Analyze edge patterns
analyze_edge_patterns(X_train, y_train, emotion_map)
```

### Create and Analyze Histogram Features

```python
def analyze_histogram_features(X_train, y_train, emotion_map):
    """
    Creates and analyzes histogram features for emotion classification.
    """
    # Get unique emotions
    unique_emotions = sorted(np.unique(y_train))
    
    # Compute histograms for each image
    n_bins = 32
    hist_features = []
    
    for img in X_train:
        hist, _ = np.histogram(img, bins=n_bins, range=(0, 255))
        hist = hist / np.sum(hist)  # Normalize
        hist_features.append(hist)
    
    hist_features = np.array(hist_features)
    
    # Compute average histogram for each emotion
    avg_histograms = {}
    for emotion_idx in unique_emotions:
        emotion_indices = np.where(y_train == emotion_idx)[0]
        emotion_hists = hist_features[emotion_indices]
        avg_histograms[emotion_idx] = np.mean(emotion_hists, axis=0)
    
    # Visualize average histograms
    plt.figure(figsize=(12, 8))
    
    for emotion_idx in unique_emotions:
        plt.plot(avg_histograms[emotion_idx], label=emotion_map[emotion_idx])
    
    plt.title('Average Intensity Histograms by Emotion')
    plt.xlabel('Intensity Bin')
    plt.ylabel('Normalized Frequency')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    # Analyze histogram differences
    plt.figure(figsize=(12, 8))
    
    # Compute histogram differences
    hist_diffs = {}
    for i, emotion1 in enumerate(unique_emotions):
        for emotion2 in unique_emotions[i+1:]:
            diff = np.abs(avg_histograms[emotion1] - avg_histograms[emotion2])
            hist_diffs[(emotion1, emotion2)] = diff
            
            plt.plot(
                diff, 
                label=f'{emotion_map[emotion1]} vs {emotion_map[emotion2]}'
            )
    
    plt.title('Histogram Differences Between Emotions')
    plt.xlabel('Intensity Bin')
    plt.ylabel('Absolute Difference')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    # Compute and visualize cumulative histograms
    cum_histograms = {}
    for emotion_idx in unique_emotions:
        cum_histograms[emotion_idx] = np.cumsum(avg_histograms[emotion_idx])
    
    plt.figure(figsize=(12, 8))
    
    for emotion_idx in unique_emotions:
        plt.plot(cum_histograms[emotion_idx], label=emotion_map[emotion_idx])
    
    plt.title('Cumulative Intensity Distributions by Emotion')
    plt.xlabel('Intensity Bin')
    plt.ylabel('Cumulative Frequency')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    return hist_features, avg_histograms

# Analyze histogram features
hist_features, avg_histograms = analyze_histogram_features(X_train, y_train, emotion_map)
```

###  Analyze Class Separability and Confusion Metrics

```python
from sklearn.metrics import pairwise_distances
from sklearn.neighbors import KNeighborsClassifier
```

```python
def analyze_class_separability(X_train, y_train, emotion_map):
    """
    Analyzes separability between emotion classes.
    """
    # Get unique emotions
    unique_emotions = sorted(np.unique(y_train))
    
    # Flatten images
    X_flat = X_train.reshape(X_train.shape[0], -1)
    
    # Compute class means
    class_means = {}
    for emotion_idx in unique_emotions:
        emotion_indices = np.where(y_train == emotion_idx)[0]
        class_means[emotion_idx] = np.mean(X_flat[emotion_indices], axis=0)
    
    # Compute between-class distances
    distances = np.zeros((len(unique_emotions), len(unique_emotions)))
    
    for i, emotion1 in enumerate(unique_emotions):
        for j, emotion2 in enumerate(unique_emotions):
            distances[i, j] = np.sqrt(np.sum((class_means[emotion1] - class_means[emotion2])**2))
    
    # Visualize distance matrix
    plt.figure(figsize=(10, 8))
    
    sns.heatmap(
        distances,
        annot=True,
        fmt='.1f',
        xticklabels=[emotion_map[e] for e in unique_emotions],
        yticklabels=[emotion_map[e] for e in unique_emotions],
        cmap='viridis'
    )
    
    plt.title('Euclidean Distance Between Emotion Class Means')
    plt.tight_layout()
    plt.show()
    
    # Analyze nearest neighbor confusion
    print("Analyzing k-Nearest Neighbor Confusion Patterns...")
    
    # Train a KNN classifier
    knn = KNeighborsClassifier(n_neighbors=5)
    knn.fit(X_flat, y_train)
    
    # Predict on training set (to analyze confusion patterns)
    y_pred = knn.predict(X_flat)
    
    # Compute confusion matrix
    conf_matrix = np.zeros((len(unique_emotions), len(unique_emotions)), dtype=int)
    
    for true_idx, pred_idx in zip(y_train, y_pred):
        # Get indices in unique_emotions list
        true_pos = unique_emotions.index(true_idx)
        pred_pos = unique_emotions.index(pred_idx)
        
        # Increment confusion matrix
        conf_matrix[true_pos, pred_pos] += 1
    
    # Normalize by row (true labels)
    conf_matrix_norm = conf_matrix / conf_matrix.sum(axis=1, keepdims=True)
    
    # Visualize confusion matrix
    plt.figure(figsize=(10, 8))
    
    sns.heatmap(
        conf_matrix_norm,
        annot=True,
        fmt='.2f',
        xticklabels=[emotion_map[e] for e in unique_emotions],
        yticklabels=[emotion_map[e] for e in unique_emotions],
        cmap='Blues'
    )
    
    plt.title('k-NN Confusion Matrix (Training Set)')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.tight_layout()
    plt.show()
    
    # Find most commonly confused pairs
    confusion_pairs = []
    
    for i, emotion1 in enumerate(unique_emotions):
        for j, emotion2 in enumerate(unique_emotions):
            if i != j:
                confusion_pairs.append((
                    emotion1,
                    emotion2,
                    conf_matrix_norm[i, j]
                ))
    
    # Sort by confusion rate (descending)
    confusion_pairs.sort(key=lambda x: x[2], reverse=True)
    
    # Print top confused pairs
    print("\nTop Confused Emotion Pairs:")
    for emotion1, emotion2, rate in confusion_pairs[:5]:
        print(f"- {emotion_map[emotion1]} confused as {emotion_map[emotion2]}: {rate:.2%}")
    
    return distances, conf_matrix_norm

# Analyze class separability
class_distances, confusion_matrix_norm = analyze_class_separability(X_train, y_train, emotion_map)
```

### Analyze Facial Symmetry

```python
def analyze_facial_symmetry(X_train, y_train, emotion_map, sample_size=3):
    """
    Analyzes facial symmetry for different emotions.
    """
    # Get unique emotions
    unique_emotions = sorted(np.unique(y_train))
    
    # Create figure
    plt.figure(figsize=(15, len(unique_emotions)*2.5))
    
    # For each emotion
    for i, emotion_idx in enumerate(unique_emotions):
        # Get images for this emotion
        emotion_indices = np.where(y_train == emotion_idx)[0]
        
        # Select sample images
        selected_indices = np.random.choice(emotion_indices, min(sample_size, len(emotion_indices)), replace=False)
        
        # For each sample
        for j, idx in enumerate(selected_indices):
            img = X_train[idx]
            
            # Create flipped version
            flipped = np.fliplr(img)
            
            # Calculate difference between original and flipped
            diff = np.abs(img - flipped)
            
            # Calculate symmetry score (lower means more symmetric)
            symmetry_score = np.mean(diff)
            
            # Plot original image
            plt.subplot(len(unique_emotions), sample_size*3, i*sample_size*3 + j*3 + 1)
            plt.imshow(img, cmap='gray')
            if j == 0:
                plt.ylabel(emotion_map[emotion_idx], fontsize=12)
            if i == 0:
                plt.title('Original')
            plt.axis('off')
            
            # Plot flipped image
            plt.subplot(len(unique_emotions), sample_size*3, i*sample_size*3 + j*3 + 2)
            plt.imshow(flipped, cmap='gray')
            if i == 0:
                plt.title('Flipped')
            plt.axis('off')
            
            # Plot difference
            plt.subplot(len(unique_emotions), sample_size*3, i*sample_size*3 + j*3 + 3)
            plt.imshow(diff, cmap='hot')
            plt.title(f'Diff: {symmetry_score:.2f}')
            plt.axis('off')
    
    plt.suptitle('Facial Symmetry Analysis by Emotion', y=0.98, fontsize=16)
    plt.tight_layout()
    plt.subplots_adjust(top=0.95)
    plt.show()
    
    # Compute average symmetry scores for each emotion
    symmetry_scores = {emotion_idx: [] for emotion_idx in unique_emotions}
    
    for i in range(len(X_train)):
        img = X_train[i]
        flipped = np.fliplr(img)
        diff = np.abs(img - flipped)
        symmetry_score = np.mean(diff)
        
        emotion_idx = y_train[i]
        symmetry_scores[emotion_idx].append(symmetry_score)
    
    # Visualize symmetry distributions
    plt.figure(figsize=(12, 6))
    
    for emotion_idx in unique_emotions:
        sns.kdeplot(symmetry_scores[emotion_idx], label=emotion_map[emotion_idx])
    
    plt.title('Facial Symmetry Score Distributions by Emotion')
    plt.xlabel('Symmetry Score (lower = more symmetric)')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    # Calculate average symmetry for each emotion
    avg_symmetry = {emotion_idx: np.mean(scores) for emotion_idx, scores in symmetry_scores.items()}
    
    # Visualize average symmetry
    plt.figure(figsize=(10, 6))
    
    # Sort by symmetry score
    sorted_emotions = sorted(avg_symmetry.items(), key=lambda x: x[1])
    emotions = [emotion_map[e] for e, _ in sorted_emotions]
    scores = [s for _, s in sorted_emotions]
    
    # Plot
    sns.barplot(x=emotions, y=scores)
    plt.title('Average Facial Symmetry by Emotion')
    plt.xlabel('Emotion')
    plt.ylabel('Symmetry Score (lower = more symmetric)')
    plt.xticks(rotation=45)
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    return symmetry_scores, avg_symmetry

# Analyze facial symmetry
symmetry_scores, avg_symmetry = analyze_facial_symmetry(X_train, y_train, emotion_map)
```

### Regional Feature Extraction

```python
def extract_regional_features(X_train, y_train, emotion_map):
    """
    Extracts and analyzes features from different facial regions.
    """
    # Define facial regions (approximate)
    regions = {
        'left_eye': (12, 24, 12, 24),
        'right_eye': (12, 24, 24, 36),
        'mouth': (30, 42, 18, 30),
        'nose': (24, 30, 18, 30),
        'forehead': (0, 12, 12, 36),
        'left_cheek': (24, 36, 0, 18),
        'right_cheek': (24, 36, 30, 48)
    }
    
    # Create feature set
    regional_features = {}
    
    # For each image
    for i in range(len(X_train)):
        img = X_train[i]
        emotion_idx = y_train[i]
        
        # Initialize features for this image
        if i not in regional_features:
            regional_features[i] = {
                'emotion': emotion_idx,
                'regions': {}
            }
        
        # Extract features for each region
        for region_name, (y_start, y_end, x_start, x_end) in regions.items():
            # Extract region
            region = img[y_start:y_end, x_start:x_end]
            
            # Calculate features
            features = {
                'mean': np.mean(region),
                'std': np.std(region),
                'min': np.min(region),
                'max': np.max(region),
                'range': np.max(region) - np.min(region)
            }
            
            # Store features
            regional_features[i]['regions'][region_name] = features
    
    # Visualize regions on sample images
    plt.figure(figsize=(15, 10))
    
    # Select one sample from each emotion
    samples = []
    for emotion_idx in sorted(np.unique(y_train)):
        emotion_indices = np.where(y_train == emotion_idx)[0]
        samples.append(np.random.choice(emotion_indices))
    
    # Plot each sample with regions
    for i, idx in enumerate(samples):
        img = X_train[idx]
        emotion = emotion_map[y_train[idx]]
        
        # Create RGB version for visualization
        img_rgb = np.stack([img, img, img], axis=2) / 255.0
        
        # Add region boxes
        for region_name, (y_start, y_end, x_start, x_end) in regions.items():
            # Draw rectangle
            img_rgb[y_start:y_end, x_start:x_start+1] = [1, 0, 0]  # Red
            img_rgb[y_start:y_end, x_end-1:x_end] = [1, 0, 0]
            img_rgb[y_start:y_start+1, x_start:x_end] = [1, 0, 0]
            img_rgb[y_end-1:y_end, x_start:x_end] = [1, 0, 0]
        
        # Plot
        plt.subplot(2, 4, i+1)
        plt.imshow(img_rgb)
        plt.title(f'{emotion}')
        plt.axis('off')
    
    plt.suptitle('Facial Regions for Feature Extraction', y=0.98, fontsize=16)
    plt.tight_layout()
    plt.subplots_adjust(top=0.9)
    plt.show()
    
    # Analyze regional feature importance
    # Compute average feature values for each emotion and region
    avg_features = {emotion_idx: {region: {} for region in regions} 
                   for emotion_idx in sorted(np.unique(y_train))}
    
    for i in regional_features:
        emotion_idx = regional_features[i]['emotion']
        
        for region in regions:
            for feature, value in regional_features[i]['regions'][region].items():
                if feature not in avg_features[emotion_idx][region]:
                    avg_features[emotion_idx][region][feature] = []
                
                avg_features[emotion_idx][region][feature].append(value)
    
    # Compute means
    for emotion_idx in avg_features:
        for region in avg_features[emotion_idx]:
            for feature in avg_features[emotion_idx][region]:
                avg_features[emotion_idx][region][feature] = np.mean(avg_features[emotion_idx][region][feature])
    
    # Compute feature importance based on variance between emotions
    feature_importance = {region: {} for region in regions}
    
    for region in regions:
        for feature in ['mean', 'std', 'range']:
            # Get values for each emotion
            values = [avg_features[emotion_idx][region][feature] for emotion_idx in sorted(np.unique(y_train))]
            
            # Compute variance
            variance = np.var(values)
            
            # Store importance
            feature_importance[region][feature] = variance
    
    # Visualize feature importance
    plt.figure(figsize=(15, 8))
    
    # Flatten for visualization
    region_features = []
    importance_values = []
    
    for region in regions:
        for feature in ['mean', 'std', 'range']:
            region_features.append(f'{region}_{feature}')
            importance_values.append(feature_importance[region][feature])
    
    # Sort by importance
    sorted_idx = np.argsort(importance_values)[::-1]
    sorted_features = [region_features[i] for i in sorted_idx]
    sorted_values = [importance_values[i] for i in sorted_idx]
    
    # Plot top features
    top_n = 15
    plt.bar(sorted_features[:top_n], sorted_values[:top_n])
    plt.title(f'Top {top_n} Regional Feature Importance for Emotion Classification')
    plt.xlabel('Region_Feature')
    plt.ylabel('Variance between Emotions')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    return regional_features, feature_importance

# Extract and analyze regional features
regional_features, regional_importance = extract_regional_features(X_train, y_train, emotion_map)
```
